```python
densidad_type = {'Denso':0,'Medio':1, 'Muy Ralo':2, 'Ralo':3}
estado_types = {'Bueno':0,'Regular':1, 'Malo':2, 'Suprimido':4}
riesgo_types = {'Moderada':0,'Ninguna':1, 'Severa':2}

dF['DENSIDAD'] = dF['DENSIDAD'].map(densidad_type).astype(int) 
dF['ESTADO_COPA'] = dF['ESTADO_COPA'].map(estado_types).astype(int)  
dF['ESTADO_FUSTE'] = dF['ESTADO_FUSTE'].map(estado_types).astype(int) 
dF['EST_RAIZ_GEN'] = dF['EST_RAIZ_GEN'].map(estado_types).astype(int)  
dF['ESTADO_FITO'] = dF['ESTADO_FITO'].map(estado_types).astype(int)  
dF['NIVEL_RIESGO'] = dF['NIVEL_RIESGO'].map(riesgo_types).astype(int) 
```

```python
Y_train = dF['TRATAMIENTO']
features = ['ALTURATOTAL', 'DENSIDAD', 'ESTADO_COPA', 'ESTADO_FUSTE', 'EST_RAIZ_GEN', 'NIVEL_RIESGO', 'ESTADO_FITO']
X_train = dF[features]

print(Y_train.shape,X_train.shape)
```

Modelos

```python
#REGRESION LOGISTICA
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter=700)
logreg.fit(X_train, Y_train)
```

```python
# entrenando modelo arboles de decisión
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix

def conf_mat_acc(modelo, X_train, Y_train):
    # Calcular la matriz de confusión
    y_pred = modelo.predict(X_train)
    cm = confusion_matrix(Y_train, y_pred)

    # Mostrar la matriz de confusión como gráfico
    plt.figure(figsize=(6, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()
    tick_marks = np.arange(len(set(Y_train)))
    plt.xticks(tick_marks, tick_marks)
    plt.yticks(tick_marks, tick_marks)

    # Añadir etiquetas a las celdas
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, format(cm[i, j], 'd'),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > cm.max() / 2 else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()

    # Calcular la exactitud
    true_pred = cm[0, 0] + cm[1, 1]
    total_data = np.sum(cm)
    accuracy = true_pred / total_data

    print('accuracy:', np.round(accuracy, 2))
    plt.show()
```

```python
conf_mat_acc(logreg,X_train, Y_train)
```

```python
conf_mat_acc(decision_tree,X_train, Y_train)
```
